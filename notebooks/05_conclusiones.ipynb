1. Hallazgos Principales del Análisis Exploratorio (EDA)
El análisis exploratorio univariado y bivariado demostró de manera robusta que el comportamiento de consumo en minutos mensuales (monthly_watch_time_mins) está directamente condicionado por la estructura del Plan de Suscripción contratado. Los usuarios bajo los esquemas Premium y Estándar exhiben los estratos de mayor actividad y retención de consumo en la plataforma.

Por el contrario, variables de perfil demográfico tradicional como la Edad de los suscriptores no muestran una correlación lineal ni agrupamientos estadísticamente significativos respecto al volumen de visualización o a la tasa de incidencias de soporte técnico en análisis bivariados simples. Las variables numéricas demográficas operan de forma independiente a la intensidad de uso.

2. El Impacto de la Limpieza y Validación de Datos (Clase 6 y 7)
La etapa de inspección inicial y preparación consciente fue un factor crítico para el éxito del proyecto. De no haberse implementado soluciones basadas en evidencia para las anomalías detectadas, las conclusiones de negocio y los modelos matemáticos habrían sufrido sesgos severos:

Consistencia Categórica: La unificación de cadenas de texto en subscription_plan (eliminando variantes como Basic, Std, Standard, Premiun y alteraciones de mayúsculas) impidió la fragmentación artificial de la base de datos, permitiendo segmentaciones precisas en el EDA y en los filtros de la aplicación de Streamlit.

Tratamiento de Outliers: El reemplazo mediante medianas robustas para los valores atípicos imposibles (el registro biológico de 150 años en la edad y el error de carga de 99 tickets en soporte) protegió la integridad analítica. Mantener el outlier de 150 años habría arrastrado artificialmente la media de edad y sobredimensionado la varianza global. Esto habría provocado que la Primera Componente Principal (CP1) de PCA capturara una variabilidad geométrica artificial basada en un error de carga de sistema y no en patrones reales de comportamiento.

3. Síntesis del Análisis de Componentes Principales (PCA)
Tras aplicar un escalamiento obligatorio Z-Score (StandardScaler) —indispensable para neutralizar las diferencias de magnitudes de escala entre los minutos de consumo y las unidades de tickets— el algoritmo de PCA permitió resumir la variabilidad de las tres dimensiones numéricas continuas:

CP1 (Perfil de Madurez y Asistencia): Captura de forma conjunta la variabilidad asociada a la edad del usuario y su interacción con soporte técnico (loadings altos y positivos de 0.718 y 0.693). Define un perfil analítico de usuarios maduros que demandan mayor asistencia operativa.

CP2 (Eje de Intensidad de Consumo): Dominada casi en su totalidad por la variable monthly_watch_time_mins (loading de 0.900), aislando el volumen neto de reproducción como un vector de segmentación independiente.

La combinación de las dos primeras componentes logra retener la estructura esencial de la información original, permitiendo una proyección multivariada óptima y sin pérdidas críticas de varianza.

4. Limitaciones del Proyecto y Mejoras Futuras
Limitaciones Identificadas:
El alcance de las conclusiones obtenidas se encuentra condicionado de forma estricta por la naturaleza transversal del dataset provisto y por el número limitado de variables operativas. Al tratarse de una captura estática de usuarios y no contar con un histórico temporal de comportamiento, logs de navegación detallados por sesión o series de tiempo, no es factible evaluar la causalidad directa de la baja de usuarios (churn) ni registrar variaciones estacionales en el consumo de contenido.

Mejoras Futuras Propuestas:

Métricas de Retención: Incorporar variables contextuales relativas al costo de adquisición del cliente (CAC), tasa de cancelación (churn rate) e interacciones cualitativas de soporte clasificadas por motivo técnico/administrativo.

Automatización de Reglas de Calidad: Implementar un pipeline automatizado de validación continua (vía expresiones regulares o esquemas de datos) en la interfaz de registro de la plataforma de streaming para impedir, en origen, el ingreso de cadenas inconsistentes o edades biológicamente inverosímiles.
